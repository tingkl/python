```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
# plt.xlim(-5, 15)
# plt.ylim(0, 1.5)
# plt.axis([-1, 11, -2, 2])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title('Welcome to the ML World!')
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
```
